```python
# Assistant
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the data with proper path handling
df = pd.read_csv(r'C:\Users\duong\Stat Group 1\Final_clean_table_STATS_TRUST_X_VACC.csv')

# Display the first few rows to verify the data was loaded correctly
print("Dataset preview:")
print(df.head())

# Step 1: Calculate Mean
trust_mean = np.mean(df['Trust_rate'])
vacc_mean = np.mean(df['Vacc_rate'])
print("\nStep 1: Calculate Mean")
print(f"Mean of Trust_rate: {trust_mean:.4f}")
print(f"Mean of Vacc_rate: {vacc_mean:.4f}")

# Step 2: Calculate Variance
trust_variance = np.var(df['Trust_rate'], ddof=1)  # ddof=1 for sample variance
vacc_variance = np.var(df['Vacc_rate'], ddof=1)
print("\nStep 2: Calculate Variance")
print(f"Variance of Trust_rate: {trust_variance:.4f}")
print(f"Variance of Vacc_rate: {vacc_variance:.4f}")

# Step 3: Calculate Standard Deviation
trust_std = np.std(df['Trust_rate'], ddof=1)  # ddof=1 for sample standard deviation
vacc_std = np.std(df['Vacc_rate'], ddof=1)
print("\nStep 3: Calculate Standard Deviation")
print(f"Standard Deviation of Trust_rate: {trust_std:.4f}")
print(f"Standard Deviation of Vacc_rate: {vacc_std:.4f}")

# Step 4: Calculate Skewness (Fisher-Pearson adjusted moment coefficient)
n = len(df)

# For Trust_rate
trust_diff = df['Trust_rate'] - trust_mean
trust_skewness_numerator = (np.sum((trust_diff/trust_std)**3))*n
trust_skewness_denominator = (n-1)*(n-2)
trust_adjusted_skewness = trust_skewness_numerator / trust_skewness_denominator

# For Vacc_rate
vacc_diff = df['Vacc_rate'] - vacc_mean
vacc_skewness_numerator = (np.sum((vacc_diff/vacc_std)**3))*n
vacc_skewness_denominator = (n-1)*(n-2)
vacc_adjusted_skewness = vacc_skewness_numerator / vacc_skewness_denominator


print("\nStep 4: Calculate Skewness (Fisher-Pearson adjusted moment coefficient)")
print(f"Adjusted skewness of Trust_rate: {trust_adjusted_skewness:.4f}")
print(f"Adjusted Skewness of Vacc_rate: {vacc_adjusted_skewness:.4f}")

# Step 5: Calculate Covariance
covariance = np.sum(trust_diff * vacc_diff) / (n - 1)  # Sample covariance
print("\nStep 5: Calculate Covariance")
print(f"Covariance between Trust_rate and Vacc_rate: {covariance:.4f}")

# Step 6: Calculate Correlation Coefficient
correlation = covariance / (trust_std * vacc_std)
print("\nStep 6: Calculate Correlation Coefficient")
print(f"Correlation between Trust_rate and Vacc_rate: {correlation:.4f}")

# Step 7: Linear Regression
m = covariance / trust_variance  # Slope
b = vacc_mean - m * trust_mean   # Intercept
print("\nStep 7: Calculate Linear Regression")
print(f"Linear Regression Equation: Vacc_rate = {m:.4f} * Trust_rate + {b:.4f}")

# Create a figure with multiple subplots
plt.figure(figsize=(15, 12))

# Plot 1: Distribution of Trust_rate with skewness
plt.subplot(2, 2, 1)
sns.histplot(df['Trust_rate'], kde=True, color='blue')
plt.axvline(trust_mean, color='red', linestyle='--', label=f'Mean: {trust_mean:.4f}')
plt.title(f'Distribution of Trust Rate\nSkewness: {trust_skewness_adjusted:.4f}')
plt.xlabel('Trust Rate')
plt.ylabel('Frequency')
plt.legend()

# Plot 2: Distribution of Vacc_rate with skewness
plt.subplot(2, 2, 2)
sns.histplot(df['Vacc_rate'], kde=True, color='green')
plt.axvline(vacc_mean, color='red', linestyle='--', label=f'Mean: {vacc_mean:.4f}')
plt.title(f'Distribution of Vaccination Rate\nSkewness: {vacc_skewness_adjusted:.4f}')
plt.xlabel('Vaccination Rate')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()

# Plot the scatter plot with regression line
plt.figure(figsize=(10, 6))
plt.scatter(df['Trust_rate'], df['Vacc_rate'], alpha=0.7)
plt.title('Relationship between Trust Rate and Vaccination Rate')
plt.xlabel('Trust Rate')
plt.ylabel('Vaccination Rate')
plt.grid(True, alpha=0.3)

# Add regression line
x_range = np.linspace(min(df['Trust_rate']), max(df['Trust_rate']), 100)
y_pred = m * x_range + b
plt.plot(x_range, y_pred, "r--", alpha=0.8, label=f'y = {m:.4f}x + {b:.4f}')
plt.legend()

# Add correlation coefficient and skewness to the plot
plt.annotate(f'Correlation: {correlation:.4f}\n'
             f'Trust Skewness: {trust_skewness_adjusted:.4f}\n'
             f'Vacc Skewness: {vacc_skewness_adjusted:.4f}', 
             xy=(0.05, 0.95), 
             xycoords='axes fraction', 
             fontsize=10,
             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.7))
plt.show()
```
